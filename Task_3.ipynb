## State Wise Analysis ##

```python
ps = sales.groupby('State')[['Sales','Profit']].sum().sort_values(by='Sales',ascending=False)
ps[:].plot.bar(color=['blue','orange'],figsize=(15,8))
plt.title('Profit/Loss and Sales across states')
plt.xlabel('States')
plt.ylabel('Profit/Loss and Sales')
plt.show()
```

### California has highest profit and Texas has lowest profit
### California has highest sales and Texas has lowest North Dekota

## Region Wise Analysis ##

```python
sales.groupby('Region')[['Profit','Sales']].sum().sort_values(by='Sales',ascending=False).plot.bar(color=['orange','maroon'],figsize=(10,8))
plt.ylabel('Profit/Loss and Sales')
plt.show()
```

### West has highest profit and Central has lowest profit
### West has highest sales and South has lowest sales

## City wise analysis

```python
g = sales.groupby('City')[['Profit','Sales']].sum().sort_values(by='Sales',ascending=False)
g.plot.bar(color=['orange','maroon'],figsize=(25,18))
```

```python
g.max
```

```python
g.min()
```

### New york City has Highest profit and sales.
### Philadelphia has least profit.
### Abilene has lowest sales.

# Ship-Mode Counts

```python
sns.countplot(sales['Ship Mode'])
plt.show()
```

### Standard class ship-mode is highest use by the customers.

# Profit with respect to category

```python
plt.title("Percentage Profit with respect to Category")
plt.pie(sales['Profit'].groupby(sales['Category']).sum(),labels=sales['Profit'].groupby(sales['Category']).sum().index,autopct='%0.1f%%')
plt.show()
```

### Technology category has highest profit on the other hand Furniture has lowest.

# Profit with respect to Sub-Category

```python
plt.figure(figsize=(15,7))
plt.title("Profit vs Sub-Category ")
sns.barplot(x='Sub-Category',y='Profit',data=sales)
plt.show()
```

```python
plt.figure(figsize=(18,8))
sns.countplot(sales['Sub-Category'])
plt.show()
```

### Copiers sub-category has highest profit whereas tables has lowest.
### Binders sub-category has highest count whereas Copiers has lowest.

# Profit with respect to Discount

```python
sns.barplot(data=sales, x="Discount", y="Profit")
plt.show()
```

### If Seller give 0.1% discount then they will get highest profit however, if seller give more than 0.1% discount then they will be in loss.

# Profit with respect to segment


```python
plt.title("%Profit with segment")
plt.pie(sales['Profit'].groupby(sales['Segment']).sum(),labels=sales['Profit'].groupby(sales['Segment']).sum().index,autopct='%0.1f%%')
plt.show()
```

### Consumer segment has more profit comapare to other segments.
